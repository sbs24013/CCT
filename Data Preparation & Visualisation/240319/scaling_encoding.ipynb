**MinMaxScaler:**

사용 시기: MinMaxScaler는 데이터를 고정된 범위(일반적으로 0과 1 사이)로 조정합니다. 원래 분포의 모양을 유지하고 데이터의 최소값과 최대값을 유지하려는 경우에 적합합니다. 왜곡된 데이터에 최소값과 최대값 측면에서 의미 있는 해석이 있는 경우 MinMaxScaler가 적합할 수 있습니다.

기울어진 데이터에 대한 영향: MinMaxScaler는 특히 왜곡이 극단적인 값으로 인해 발생하는 경우 이상값의 영향을 받을 수 있습니다. 치우친 분포에 이상값이 포함된 경우 MinMaxScaler는 최소값과 최대값을 기준으로 데이터 크기를 조정하여 대부분의 데이터 포인트를 범위 중심으로 압축하므로 최선의 선택이 아닐 수 있습니다.

**정규화 스케일러(StandardScaler):**

사용 시기: StandardScaler와 같은 정규화 스케일링은 평균이 0, 표준편차가 1이 되도록 데이터를 스케일링합니다. 데이터 분포를 표준화하고 평균을 제거하려는 경우에 적합합니다. 분포의 형태를 유지합니다. 치우친 데이터가 평균을 중심으로 하고 일관된 척도를 가져야 하는 경우 정규화 척도가 적절할 수 있습니다.

편향된 데이터에 미치는 영향: 정규화 스케일러는 MinMaxScaler에 비해 이상값에 덜 민감합니다. 평균 및 표준 편차를 기반으로 데이터의 규모를 조정하여 왜도가 있는 극단값에 대해 더욱 견고하게 만듭니다. 그러나 왜도가 극단적이거나 이상값이 많은 경우에는 정규화 스케일링으로 문제를 완전히 해결하지 못할 수도 있습니다.

```python
import sklearn.preprocessing as preproc

df[" n_tokens_content"]
```

```python
# MinMax Scalier
# Skewed
# 최소값과 최대값이 의미가 있는 경우
# 편향된 데이터를 그대로 유지

df["min_max"] = preproc.minmax_scale(df[[" n_tokens_content"]])
```

```python
df["min_max"]
```

```python
# Normalized data
# ??
# 평균을 제거
# 분포를 그대로 유지
# 편향된 데이터를 중심으로 옮겨야 하는 경우

df["standardized"] = preproc.StandardScaler().fit_transform(df[[" n_tokens_content"]])
```

```python
df["standardized"]
```

```python
# Sparsed Data
# L2
# 데이터와 훈련 세트 사이의 거리를 계산하여 예측값을 사용

df["l2_normalized"] = preproc.normalize(df[[" n_tokens_content"]], axis = 0)

```
